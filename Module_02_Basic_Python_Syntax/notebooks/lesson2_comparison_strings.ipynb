# Comparison Operators with Strings

This notebook explores how Python compares strings using comparison operators and Unicode/lexicographical order.

## 1. Equality (`==`) and Not Equal (`!=`) with Strings

```python
print("a string" == "a string")  # True
print("Hello" == "hello")        # False (case-sensitive)
print("cat" != "dog")            # True
print("4 + 5" == 4 + 5)          # False (string vs integer)
```

## 2. Lexicographical Order with `>` and `<`
Python compares strings character by character using their Unicode code points.

```python
print("Wednesday" > "Friday")  # True ('W' > 'F')
print("apple" < "banana")          # True
print("Zoo" > "apple")             # False ('Z' < 'a')
```

## 3. `>=` and `<=` with Strings

```python
print("my computer" >= "my chair")
print("Spring" <= "Winter")
print("pineapple" >= "pineapple")
```

## Summary
- String comparison is **case-sensitive** and based on Unicode values.
- `==` and `!=` check if two strings are identical or different.
- `>`, `<`, `>=`, and `<=` compare strings in lexicographical (dictionary-like) order.

# Comparison Operators with Numbers

This notebook covers Python comparison operators with numeric values (`int` and `float`) and how they return Boolean results (`True` or `False`).

## 1. Equality (`==`) and Not Equal (`!=`)
Use these to check if two values are the same or different.

```python
# Equality examples
print(32 == 30 + 2)      # True
print(5 + 10 == 6 + 7)   # False

# Not equal examples
print(10 != 5)           # True
print(3 * 4 != 12)       # False
```

## 2. Assignment (`=`) vs Equality (`==`)
Remember that `=` assigns a value to a variable, while `==` compares two values.

```python
# Assignment vs equality
my_variable = 3 * 5
print(my_variable)       # 15

# This is a comparison
print(my_variable == 15) # True
```

## 3. Greater Than (`>`) and Less Than (`<`)

```python
print(11 > 3 * 3)   # True
print(4 / 2 > 8 - 4)   # False

print(2 < 10)          # True
print(10 < 2)          # False
```

## 4. Greater Than or Equal (`>=`) and Less Than or Equal (`<=`)

```python
print(12 * 2 >= 24)   # True (24 >= 24)
print(18 / 2 >= 15)   # False (9 is not >= 15)

print(5 <= 5)         # True
print(7 <= 3)         # False
```

## Summary
- Comparison operators (`==`, `!=`, `>`, `<`, `>=`, `<=`) compare values and return `True` or `False`.
- `=` is used for assignment, not for comparison.
- These operators are the foundation for conditionals and loops in Python.

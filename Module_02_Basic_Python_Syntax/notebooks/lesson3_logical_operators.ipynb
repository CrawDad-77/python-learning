# Logical Operators in Python

This notebook covers the logical operators `and`, `or`, and `not`, and how they combine comparison expressions.

## 1. The `and` Operator
Returns `True` only if **both** sides are `True`.

```python
print((6*3 >= 18) and (9+9 <= 36/2))  # True and True → True

print("Nairobi" < "Milan" and "Nairobi" > "Hanoi")
# First comparison is False, second is True → overall result False
```

## 2. The `or` Operator
Returns `True` if **at least one** side is `True`.

```python
country = "United States"
city = "New York City"

# True or True → True
print((15/3 < 2+4) or (0 >= 6-7))

# False or True → True
print(country == "New York City" or city == "New York City")
```

## 3. The `not` Operator
Inverts a Boolean value: `not True` → False, `not False` → True.

```python
x = 2*3 > 6    # 6 > 6 → False
print("x:", x)
print("not x:", not x)

today = "Monday"
print(not today == "Tuesday")  # not (False) → True
```

## Summary
- `and`: all conditions must be `True`.
- `or`: at least one condition must be `True`.
- `not`: flips a Boolean value.
- Logical operators are typically used together with comparison operators to build complex conditions.

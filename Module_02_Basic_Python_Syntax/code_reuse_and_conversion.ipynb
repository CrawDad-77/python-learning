# Code Reuse: Using Functions to Avoid Repeating Work
This notebook explains how functions help you reuse code, plus an example of a function that converts fluid ounces to milliliters.

## 🚫 The Long Way (Repeating Code)

```python
name = "Kay"
number = len(name) * 9
print("Hello " + name + ". Your lucky number is " + str(number))

name = "Cameron"
number = len(name) * 9
print("Hello " + name + ". Your lucky number is " + str(number))
```

## ⭐ The Better Way — Using a Function

```python
def lucky_number(name):
    number = len(name) * 9
    print("Hello " + name + ". Your lucky number is " + str(number))

lucky_number("Kay")
lucky_number("Cameron")
```

# 📏 Converting Fluid Ounces to Milliliters
Here we create a function that *returns* a value instead of just printing it.

```python
# This function converts fluid ounces to milliliters and returns the result.
def convert_volume(fluid_ounce):
    ml = fluid_ounce * 29.5
    return ml
```

### Using the function inside print()

```python
print("The volume in milliliters is " + str(convert_volume(2)))
```

### Reusing the function and doubling the input

```python
print("The volume in milliliters is " + str(convert_volume(2) * 2))
```

### Equivalent alternative calculation

```python
# This produces the same result:
print("The volume in milliliters is " + str(convert_volume(4)))
```

# The Principles of Code Reuse
This notebook explains how repeating code can be replaced with **functions**, making your programs cleaner and easier.

Think of a *function* like a **magic machine**: you give it something, and it does all the work for you!

## 🚫 The Long Way (Repeating Code)

```python
name = "Kay"
number = len(name) * 9
print("Hello " + name + ". Your lucky number is " + str(number))

name = "Cameron"
number = len(name) * 9
print("Hello " + name + ". Your lucky number is " + str(number))
```

### ❗ What's happening here?
You're doing the **same work twice** — only the names change!

## ⭐ The Better Way — Using a Function

```python
def lucky_number(name):
    number = len(name) * 9
    print("Hello " + name + ". Your lucky number is " + str(number))

lucky_number("Kay")
lucky_number("Cameron")
```

### 🎉 Why this is awesome
- You build the *machine* one time.
- You can reuse it again and again.
- No more repeating the same code!

Just change the input name, and the function does the work.

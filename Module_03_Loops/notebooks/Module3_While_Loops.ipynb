# Module 3 — While Loops

```python
x = 0
while x < 5:
    print('Not there yet, x=' + str(x))
    x += 1
print('x=' + str(x))
```

```python
def attempts(n):
    x = 1
    while x <= n:
        print('Attempt ' + str(x))
        x += 1
    print('Done')

attempts(5)
```

```python
def count_down(start_number):
    current = start_number
    while current > 0:
        print(current)
        current -= 1
    print('Zero!')

count_down(3)
```

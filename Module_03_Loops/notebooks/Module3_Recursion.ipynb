# Module 3 — Recursion

```python
def factorial(n):
    if n < 2:
        return 1
    return n * factorial(n-1)

factorial(4)
```

```python
def factorial_trace(n):
    print('Called with', n)
    if n < 2:
        return 1
    result = n * factorial_trace(n-1)
    print('Returning', result)
    return result

factorial_trace(4)
```

# Module 3 — Nested For Loops

```python
for left in range(7):
    for right in range(left, 7):
        print('[' + str(left) + '|' + str(right) + ']', end=' ')
    print()
```

```python
teams = ['Dragons', 'Wolves', 'Pandas', 'Unicorns']
for home in teams:
    for away in teams:
        if home != away:
            print(home + ' vs ' + away)
```

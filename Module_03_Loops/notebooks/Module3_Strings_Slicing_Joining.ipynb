# Module 3 — Strings, Slicing, Joining

```python
greeting = 'Hello'
for c in greeting:
    print(c)
```

```python
text = 'Programming'
print(text[:6])
print(text[7:])
print(text[-3:])
```

```python
print('Hello' + ' ' + 'world')
print(' '.join(['Hello', 'world']))
```

```python
def format_phone(p):
    return '(' + p[:3] + ') ' + p[3:6] + '-' + p[-4:]

format_phone('2025551212')
```
